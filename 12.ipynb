```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


dataset = pd.read_csv('Google-Playstore.csv')

# Convert rating to numeric (if not already)
dataset['Rating'] = pd.to_numeric(dataset['Rating'], errors='coerce')

# You may want to drop rows with missing values
dataset = dataset.dropna(subset=['Size', 'Rating'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Size', y='Rating', data=dataset)

# Add titles and labels
plt.title('Relationship between App Size and Rating')
plt.xlabel('Size (MB/GB)')
plt.ylabel('Rating')

# Show the plot
plt.show()
```
