```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


flight_data = pd.read_csv('Clean_Dataset.csv')

plt.figure(figsize=(10, 6))  # Adjust the figure size as necessary
sns.boxplot(x='airline', y='price', data=flight_data)
plt.xticks(rotation=45)  # Rotate the x-labels if they are crowded
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.show()
```
