```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
dataset = pd.read_csv('Google-Playstore.csv')

# Ensure the 'Price' column is treated as a string
dataset['Price'] = dataset['Price'].astype(str)

# Now, remove the dollar sign and convert to numeric
dataset['Price'] = pd.to_numeric(dataset['Price'].str.replace('$', ''), errors='coerce')

# Calculate average price by app type
average_prices = dataset.groupby('Category')['Price'].mean()

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=average_prices.index, y=average_prices.values)

# Add titles and labels
plt.title('Average App Price by Category')
plt.xlabel('Type of App')
plt.ylabel('Average Price')

# Show the plot
plt.show()
```
