```python
import pandas as pd

# Function to clean and convert 'Installs' column
def clean_installs(val):
    if pd.isna(val):
        return 0
    return int(val.replace('+', '').replace(',', ''))

# Define the columns you need
columns_to_use = ['App Name', 'Installs']

# Initialize an empty DataFrame to hold aggregated results
top_apps = pd.DataFrame(columns=columns_to_use)

# Read the file in chunks
chunk_size = 10000  # Adjust chunk size based on your system's memory
for chunk in pd.read_csv('Google-Playstore.csv', usecols=columns_to_use, chunksize=chunk_size):
    # Clean 'Installs' column
    chunk['Installs'] = chunk['Installs'].apply(clean_installs)
    # Aggregate data
    aggregated = chunk.groupby('App Name')['Installs'].sum().reset_index()
    top_apps = top_apps.append(aggregated, ignore_index=True)

# Further aggregate if same app appears in multiple chunks
final_aggregated = top_apps.groupby('App Name')['Installs'].sum().reset_index()

# Get top 10 apps by install count
top_10_apps = final_aggregated.sort_values(by='Installs', ascending=False).head(10)

# Display the top 10 apps
print(top_10_apps)
```
