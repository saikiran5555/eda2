```python
import pandas as pd


dataset = pd.read_csv('Google-Playstore.csv')

# Convert ratings to numeric (if not already)
dataset['Rating'] = pd.to_numeric(dataset['Rating'], errors='coerce')

# Drop rows with missing ratings
dataset = dataset.dropna(subset=['Rating'])


import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(15, 10))

# Create the boxplot
sns.boxplot(x='Category', y='Rating', data=dataset)

# Improve the aesthetics
plt.xticks(rotation=90)  # Rotate category labels for better readability
plt.title('App Ratings by Category')
plt.ylabel('Ratings')
plt.xlabel('Category')

# Show the plot
plt.show()
```
